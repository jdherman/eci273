Lists vs. arrays (__IMPORTANT__)

A regular python list:

```python
L = [6, 3, 7, 4, 5]
```

A numpy array:

```python
A = np.array([6, 3, 7, 4, 5])
```

`A` behaves like a matlab array.

What do you expect to happen?

```python
print(L*3) # prints the list 3 times
print(A*3) # multiplies each value in the array by 3
```

__Array Indexing__ `[start:stop:step]`

Defaults if omitted: start=0, stop=N, step=1 (N is the number of values in the array)

```python
print(A)
print(A[2:]) # from i to end
print(A[:2]) # from beginning to i (not inclusive)
print(A[-1]) # negative indexing is allowed!
print(A[::2]) # every other element (omit start and stop)
print(A[::-1]) # reverse the array (step = -1)
print(A[A < 5]) # logical indexing
```

Create a matrix of zeros:

```python
M = np.zeros((10,5))
M[:,:] = 4
print(M)
```

Vectorized functions (__IMPORTANT__):

`numpy` has many built-in functions to simplify matrix operations

```python
N = 10000000
M = np.ones((N,5))
```

Use vectorized functions whenever you can

e.g., don't do this...

```python
s = 0
for i in range(N):
    for j in range(5):
        s += M[i,j]
        
print(s)
```

Instead do this...

```python
print(M.sum())
# print(np.sum(M)) # either syntax is fine
```
